# Question 1: Which clusters of chronic conditions can be identified using these data?

## Approach 1: Clustering

```python
from sklearn.cluster import KMeans
```

### Trying different numbers of clusters: from 2 up to 6

```python
# let's run Kmeans a few times creating different numbers of clusters
nr_of_clusters = range(2,7)

clust_results = clust.copy()

for n in nr_of_clusters:
    # run Kmeans with n clusters
    km = KMeans(n_clusters=n, random_state=0, n_init=100, max_iter=10000, n_jobs = -2).fit(clust)
    # save the labels in the results dataframe
    clust_results["kmeans_nclust{}_labels".format(n)] = km.labels_
```

```python
clust_results.head()
```

### Nr of people per cluster

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "last_expr"
```

```python
# how's the distribution of patients for each run?
for n in nr_of_clusters:
    plt.hist(clust_results["kmeans_nclust{}_labels".format(n)]);
    plt.title("Nr of people per cluster in {}-cluster model".format(n))
    plt.show();
    
```

### A closer look within the clusters

```python
# let's see how many chronic diseases people have per class
clust_results["n_chronic_diseases"] = clust_results.loc[:, conditions].sum(axis=1)
clust_results.head()
```

```python
for n in nr_of_clusters:
    print("\n\n\n\n#####################  Nr of clusters: {}  #######################\n\n\n\n".format(n))
    
    
    colname = "kmeans_nclust{}_labels".format(n)
    
    
    subplot_cols = 2
    subplot_rows = n
    subplot_index = 1
    
    
    plt.hist(clust_results[colname]);
    plt.title("Nr of people per class when using {} clusters".format(n), fontsize=16)
    plt.ylabel("Nr of people in class")
    plt.xlabel("Class")
    plt.xticks(range(n))
    plt.show()
    print("\n")
    
    
    plt.figure(figsize=(10,subplot_rows*4));
    
    
    
    for klasse in sorted(set(clust_results[colname])):
        
        nr_of_people_in_class = len(clust_results.loc[clust_results[colname] == klasse, ])

        
        # nr of diseases per cluster
        plt.subplot(subplot_rows, subplot_cols, subplot_index);
        subplot_index += 1
        plt.hist(clust_results.loc[clust_results[colname] == klasse, "n_chronic_diseases"]);
        mean_nr_of_diseases = round(clust_results.loc[clust_results[colname] == klasse, "n_chronic_diseases"].mean(), 2)
        plt.title("Number of chronic diseases");
        plt.xticks(range(0,8))
        plt.xlabel("Number of chronic diseases\nMean: {}".format(mean_nr_of_diseases));
        plt.ylabel("Class {}\nN={}".format(klasse, nr_of_people_in_class));

        
        # count per disease per cluster
        plt.subplot(subplot_rows, subplot_cols, subplot_index);
        subplot_index += 1
        clust_results.loc[clust_results[colname] == klasse, conditions].sum().plot(kind="bar", color="r");
        plt.title("Count of chronic diseases for class {}".format(klasse));
        plt.ylim([0, 6000])
        # plt.grid(True);
    
    plt.tight_layout();
    plt.show();
```

## Principal Component Analysis

https://jakevdp.github.io/PythonDataScienceHandbook/05.09-principal-component-analysis.html

```python
from sklearn.decomposition import PCA

clust_pca = clust.copy()

pca = PCA(n_components=2)

pca.fit(clust_pca)
print(pca.components_)
print(pca.explained_variance_)

projected = pca.fit_transform(clust_pca)
projected.shape

plt.scatter(projected[:, 0], projected[:, 1],
            c=clust_results["kmeans_nclust2_labels"], edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('spectral', 2))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

### Choosing the number of components

A vital part of using PCA in practice is the ability to estimate how many components are needed to describe the data. This can be determined by looking at the cumulative explained variance ratio as a function of the number of components:

```python
pca = PCA().fit(clust_pca)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
# Ok, so let's go with 4 components


from sklearn.decomposition import PCA

clust_pca = clust.copy()
print(clust_pca.shape)

pca = PCA(n_components=4)

pca.fit(clust_pca)
print(pca.components_)
print(pca.explained_variance_)


projected = pca.fit_transform(clust_pca)
projected.shape

plt.scatter(projected[:, 0], projected[:, 1],
            c=clust_results["kmeans_nclust4_labels"], edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('spectral', 4))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

```python
pca_components = pd.DataFrame(pca.transform(clust_pca), columns=['PCA{}'.format(i) for i in range(4)], index=clust_pca.index)
pca_results = pd.concat([clust_pca, pca_components], axis=1)
pca_results.head()
```

```python
corrs = pca_results.corr().loc[conditions, [c for c in pca_results.columns if "PCA" in c]]

plt.figure(figsize=(5, 5))

sns.heatmap(corrs, cmap="BuGn")
```

## Let's try a few classifiers without tuning parameters

### Let's first scale the data

```python
scaler = StandardScaler()  
scaler.fit(X_train)  
X_train = scaler.transform(X_train)  
X_test = scaler.transform(X_test)
```

### ...Baseline (the dumbest classifier)

```python
# visualize the scores (as in: https://github.com/ianozsvald/data_science_delivered/blob/master/ml_creating_correct_capable_classifiers.ipynb)
def scores_hist(scores, classifier_name):
    score_line = f"Scores (Accuracy) mean={scores.mean():0.2f} +/- {scores.std():0.2f} (1 s.d.)"
    fig, ax = plt.subplots()
    pd.Series(scores).hist(ax=ax, bins=BINS)
    plt.grid(False)
    ax.set_title(f"RepeatedKFold ({len(scores)} folds) with {classifier_name}\n" + score_line);
    ax.set_xlabel("Score")
    ax.set_ylabel("Frequency");
```

```python
strategy="stratified"

dummy_clf = DummyClassifier(strategy="stratified")
dummy_clf = dummy_clf.fit(X_train, y_train)

dummy_scores = cross_val_score(dummy_clf, X, y, cv=CV_N_REPEATS, n_jobs=3, verbose=1)

scores_hist(dummy_scores, "DummyClassifier(stratified)")
```

```python
strategy="most_frequent"

dummy_mf_clf = DummyClassifier(strategy="stratified")
dummy_mf_clf = dummy_mf_clf.fit(X_train, y_train)

dummy_mf_scores = cross_val_score(dummy_mf_clf, X, y, cv=CV_N_REPEATS, n_jobs=3, verbose=1)

scores_hist(dummy_mf_scores, "DummyClassifier({strategy}")
```

### A tree

```python
tree_clf = tree.DecisionTreeClassifier(min_samples_split=100, min_samples_leaf=50)
tree_clf = tree_clf.fit(X_train, y_train)

tree_scores = cross_val_score(tree_clf, X, y, cv=CV_N_REPEATS, n_jobs=3, verbose=1)

scores_hist(tree_scores, "DecisionTreeClassifier")
```

```python
cvs = []
scores = []
max_accuracy_scores = []
mean_accuracy_scores = []

for cv in range(2, 20):
    cvs.append(cv)
    accuracy_scores = cross_val_score(tree_clf, X, y, cv=cv, n_jobs=3, verbose=1)
    scores.append(accuracy_scores)
    max_accuracy_scores.append(max(accuracy_scores))
    mean_accuracy_scores.append(np.mean(accuracy_scores))
    print("Max score for {}-fold cross-validation: {}".format(cv, max(accuracy_scores)))
    print("Mean score for {}-fold cross-validation: {}".format(cv, np.mean(accuracy_scores)))
    
plt.plot(cvs, mean_accuracy_scores)
plt.title("Nr of folds vs mean accuracy score")
plt.xlabel("Nr of folds for cross-validation")
plt.ylabel("Mean accuracy of cross-validation")
plt.show();

plt.plot(cvs, max_accuracy_scores)
plt.title("Nr of folds vs max accuracy score")
plt.xlabel("Nr of folds for cross-validation")
plt.ylabel("Max accuracy of cross-validation")
plt.show();
```

```python
dot_data = tree.export_graphviz(tree_clf, out_file=None,
                                max_depth=4,
                               feature_names=X.columns,
                               class_names=["NOT chronically ill", "Chronically ill"],
                               filled=True) 
graph = graphviz.Source(dot_data) 
graph.render("../output/tree")
```

### Compare scores

```python
def compare_scores(first_scores, first_classifier_name, second_scores, second_classifier_name, bins=10):
    fig, ax = plt.subplots()
    df = pd.DataFrame({f"{first_classifier_name}": first_scores, f"{second_classifier_name}": second_scores})
    df.plot(kind='hist', ax=ax, bins=bins)
    ax.set_xlabel("Score")
    ax.set_title(f"{first_classifier_name} vs {second_classifier_name} Scores");
```

```python
compare_scores(dummy_scores, "Dummyclassifier", tree_scores, "DecisionTreeClassifier", 20)
```

# Example of feature importances in random forest

http://scikit-learn.org/stable/auto_examples/ensemble/plot_forest_importances.html

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.ensemble import ExtraTreesClassifier

# Build a classification task using 3 informative features
X, y = make_classification(n_samples=1000,
                           n_features=10,
                           n_informative=3,
                           n_redundant=0,
                           n_repeated=0,
                           n_classes=2,
                           random_state=0,
                           shuffle=False)

# Build a forest and compute the feature importances
forest = ExtraTreesClassifier(n_estimators=250,
                              random_state=0)

forest.fit(X, y)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```
