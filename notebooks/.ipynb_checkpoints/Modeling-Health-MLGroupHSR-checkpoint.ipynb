# Question 1: Which clusters of chronic conditions can be identified using these data?

## Approach 1: Clustering

```python
from sklearn.cluster import KMeans
```

```python
# let's run Kmeans a few times creating different numbers of clusters
nr_of_clusters = range(2,7)

clust_results = clust.copy()

for n in nr_of_clusters:
    # run Kmeans with n clusters
    km = KMeans(n_clusters=n, random_state=0, n_init=100, max_iter=10000, n_jobs = -2).fit(clust)
    # save the labels in the results dataframe
    clust_results["kmeans_nclust{}_labels".format(n)] = km.labels_
```

```python
clust_results.head()
```

```python
# how's the distribution of patients for each run?
for n in nr_of_clusters:
    clust_results["kmeans_nclust{}_labels".format(n)].hist()
    plt.show()
    
```

Looks like there is always one predominant cluster with many people. Perhaps those who are not chronically ill?

```python
predictors = ["gender", "age", "education", "ppp", "mothereduc_gen", "fathereduc_gen", 
              "bmi", "smoke", "exercise", "drink"]
outcome = "ischronic"
```

```python
X = df_no_nan.loc[:, predictors]
y = df_no_nan.loc[:, outcome]
```
